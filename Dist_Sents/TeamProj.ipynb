```python
#Code to remove all hyperlinks 
def preprocess_text(tweet): 
    print("in preprocess: "+tweet)
    #Code to remove text link @Username 
    # text = re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA--F]))+','',tweet) 
    print('1 KRAKATOA: ')
    
    #Code to remove anything apart from a-z or 0-9 
    text = re.sub(r'@[a-zA-Z0-9]+', '', tweet) 
    print('2: ' + text)
    
    #Code to remove spaces apart from single spaces 
#     text = re.sub(r"[^A-Za-z0-9]","", text) 
#     print('3: ' + text)
#     text = re.sub(r'+ ',' ',text) 
#     print('4: ' + text)
    
    
    return text
```

```python
# need a new class created for the streamlistener
class CustomStreamListener(tweepy.StreamListener):
    
    def on_status(self, status):
        
        # print(status)
        # global_txt
        global tweets_list, name_list, userLocation_list, lang_list, country_list
        if status.text[:2] == 'RT':
            #print("Where are we?")
            return
        
        # clean the tweets
        try:
            print("IN the TRY: "+status.text)
            tweet = preprocess_text(status.text)
            print('after the TRY: ' + tweet)
            tweets_list.append(tweet)
            txt = txt + str(tweet) + ' '
        except:
            tweets_list.append('NONE')
        try:
            userLocation_list.append(preprocess_text(status.user.location))
        except:
            userLocation_list.append('NONE')
        try:
            country_list.append(preprocess_text(status.place.country))
        except:
            country_list.append('NONE')
        try: 
            name_list.append(preprocess_text(status.user.name))
        except:
            name_list.append('NONE')
        try:
            lang_list.append(preprocess_text(status.lang))
        except:
            lang_list.append('NONE')
        #print(tweet) 
#         txt=txt+str(tweet)+' '
        print('-'*50)
        
#         print (status.author.screen_name, status.created_at, 
#                status.text)
#         # Writing status data
#         with open('OutputStreaming.txt', 'a') as f:
#             writer = csv.writer(f)
#             writer.writerow([status.author.screen_name, status.created_at, status.text])


    def on_error(self, status_code):
        
        if status_code == 420:
            return False
        while True:
            try:
                streamingAPI = tweepy.streaming.Stream(auth, CustomStreamListener())
                streamingAPI.filter(track=[search_term])
            except IncompleteRead:
                continue
            except KeyboardInterrupt:
                break
#         print(ERROR)
#         print >> sys.stderr, 'Encountered error with status code:', status_code
#         return True # Don't kill the stream

    def on_timeout(self):
        print(TIMEOUT)
        print >> sys.stderr, 'Timeout...'
        return True # Don't kill the stream
```
